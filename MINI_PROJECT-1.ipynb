#normalization

```python
#normalization
import pandas as pd
import numpy as np
from sklearn import preprocessing

def preprocess_csv(csv_path, output_path):
    # Read CSV data into a pandas dataframe
    df = pd.read_csv(csv_path)
    #print(df.dtypes)

    # Initialize LabelEncoder for categorical columns
    lab=preprocessing.LabelEncoder()

    # Loop through each column in the dataframe
    for column in df.columns:

        if df[column].dtype == np.float64:

            # Normalize numerical columns to 0-255 range
            df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())*255

        elif df[column].dtype == bool:
            # Convert boolean columns to 0/1 values
            df[column] = df[column].astype(int)

        else:
            # Convert categorical columns to one-hot encoding using LabelEncoder
            df[column] = lab.fit_transform(df[column])
            df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())*255

    df.fillna(0,inplace = True)

    # Save preprocessed dataframe to CSV file
    df.to_csv(output_path)

# Call the function for each CSV file

for i, csv_file in enumerate(csv_files_b):
    output_path = f'/content/drive/MyDrive/Colab Notebooks/mini/miniprojectxl/csvnb/preprocessed_csv_b_{i}.csv'
    preprocess_csv(csv_file, output_path)
```

#csv to image conversion




```python
#image to csv conversion
import csv
from PIL import Image

def create_image_from_csv(csv_path, image_path):
    # Define image size and color mode
    width = 300
    height = 200
    color_mode = 'L'  # 'L' for grayscale, 'RGB' for color

    # Create a new grayscale image
    img = Image.new(color_mode, (width, height))

    # Create a pixel access object to modify the image data
    pixels = img.load()

    # Open the CSV file and read the data
    with open(csv_path, 'r') as csv_file:
        reader = csv.reader(csv_file)
        for row_idx, row in enumerate(reader):
            # Set the column containing string values
            string_value = row[0]
            string_column_index = 50
            if row_idx < height:
                pixels[string_column_index, row_idx] = 255 if string_value else 0  # 255 for white, 0 for black

            # Set the rest of the columns with float values
            for col_idx, value in enumerate(row[1:], start=1):
                col = int(col_idx * width / len(row))
                if row_idx < height and col < width:
                    try:
                        value = float(value)
                        pixel_value = int(value * 255)
                        pixels[col, row_idx] = pixel_value
                    except ValueError:
                        # Handle non-numeric values
                        pixels[col, row_idx] = 0

    # Save the image
    img.save(image_path)


# Call the function for each CSV file
for i, csv_file in enumerate(csv_files_b1):
    image_path = f'/content/drive/MyDrive/Colab Notebooks/mini/train/benign/b{i}.png'
    create_image_from_csv(csv_file, image_path)
```
